```python
import seaborn as sns
```

```python
df=sns.load_dataset('penguins')
df
```

```python
sns.stripplot( x='species', y='flipper_length_mm',data=df, hue='island')
```

```python
sns.catplot( x='species', y='flipper_length_mm',data=df, hue='island')
```

```python
df2 = sns.load_dataset('tips')
df2
```

```python
df2['total_bill'].min()
```

```python
df2.groupby('day', observed=False)['total_bill'].min()
```

```python
sns.violinplot(data=df2, x='day',y='total_bill')
```

```python
sns.violinplot(data=df2, x='day',y='total_bill',hue='sex',split='True')
```

```python
sns.catplot(data=df2, x='day',y='total_bill',hue='sex',split='True',kind='violin')
```

```python
sns.catplot(data=df2, x ='sex', y='total_bill',kind='box',col='smoker',row='time')
```
